```python
import numpy as np        # Filter all the prime numbers
import math

def prime_Numbers(n):
    if n % 2 == 0 and n > 2:
        return False
    return all(n % i for i in range(3, int(math.sqrt(n)) + 1, 2))
arr = np.arange(0,1000000)
foo = np.vectorize(prime_Numbers)
pbools = foo(arr)
primes = np.extract(pbools,arr)
primes
```
